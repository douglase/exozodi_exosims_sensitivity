# Create components from references

```python
# Create a Mask
def makemask():
    mask = np.zeros([67,67])+1e-12
    x = np.linspace(-1,1,67)
    y = x
    x,y = np.meshgrid(x,y)
    r = np.sqrt(x**2 + y**2)
    mask[r <= 0.75] = 1
    mask[r < 0.2] = 1e-12
    
    return mask

def masknan(array,replace):
    """
    Set values NMF can't solve for to some value, tends to not like values <= 0
    """
    array[np.isnan(array)] = replace
    array[array==np.inf] = replace
    array[array<=0] = replace
    
    return array

mask = makemask()
refbox = masknan(refbox,1e-10)

plt.figure()
plt.imshow(refbox[1])
plt.colorbar()
plt.show()

# Generate components for NMF subtraction
# 
# 

"""
multiplying by a large scalar can assist with convergence of lots of components
my understanding is that doing so blows up the faint background such that the algorithm can converge on a component
that describes the structure well. Included is a components file we generated for the cases with and without MUFs.
"""

# Uncomment to construct components
# components = nmf_imaging.NMFcomponents(refbox*1e10,mask=mask,oneByOne=True,maxiters=1e3,n_components=66)

# # Write Components
# hdu = fits.PrimaryHDU(data=components)
# hdul = fits.HDUList([hdu])
# hdul.writeto('OS9comps_analog.fits')

components = fits.getdata(nmfcompfile)
```

# NMF Subtraction

```python
def NMFSub(data,components,mask=None,cube=False):
    # Use around 1/3 of available components
    numcomps = 33
    
    # Preallocate 
    resultbox = np.zeros(components.shape)
    results = np.zeros(data.shape)
    result = results[0]
    for i in range(data.shape[0]):
        trg = data[i]
        
        plt.figure()
        plt.title('target #{}'.format(i))
        plt.imshow(trg)
        plt.colorbar()
        plt.show()
        
        model = nmf_imaging.NMFmodelling(trg=trg,
                                        components=components,
                                        n_components=numcomps,
                                        maxiters=1e2,
                                        mask_components=mask,
                                        cube=cube)
        
        if cube==True:
            for l in range(numcomps):
                best_frac = nmf_imaging.NMFbff(trg,model[l])
                resultbox[l,:,:] += nmf_imaging.NMFsubtraction(trg,model[l],frac=best_frac)
        else:
            best_frac = nmf_imaging.NMFbff(trg,model)
            print(best_frac)
            result = nmf_imaging.NMFsubtraction(trg,model,frac=best_frac)
            print(result)
            
            results[i] = result
    
    if cube==True:
        return resultbox
    
    else:
        return results
```

# Evaluation of speckle attenuation
Compute the median and standard deviation of the +11 and -11 cases

```python
from scipy.ndimage import rotate
import poppy

M11_wd_nc = masknan(m11_nmf_nc,0)/scal
P11_wd_nc = masknan(p11_nmf_nc,0)/scal
M11_nd_nc = masknan(m11_nmf_nd_nc,0)/scal
P11_nd_nc = masknan(p11_nmf_nd_nc,0)/scal

rM11_wd_nc = masknan(rm11_nmf_nc,0)/scal
rP11_wd_nc = masknan(rp11_nmf_nc,0)/scal
rM11_nd_nc = masknan(rm11_nmf_nd_nc,0)/scal
rP11_nd_nc = masknan(rp11_nmf_nd_nc,0)/scal

wd = M11_wd_nc + rotate(P11_wd_nc,22,axes=[1,2],reshape=False)
wd *= mask

nd = M11_nd_nc + rotate(P11_nd_nc,22,axes=[1,2],reshape=False)
nd *= mask

wdr = rM11_wd_nc + rotate(rP11_wd_nc,22,axes=[1,2],reshape=False)
wdr *= mask

ndr = rM11_nd_nc + rotate(rP11_nd_nc,22,axes=[1,2],reshape=False)
ndr *= mask

wd = masknan(wd,1e-10)
nd = masknan(nd,1e-10)

std_wd = np.std(wd,axis=0)
med_wd = np.median(wd,axis=0)

std_nd = np.std(nd,axis=0)
med_nd = np.median(nd,axis=0)

std_wd_raw = np.std(wdr,axis=0)
med_wd_raw = np.median(wdr,axis=0)

std_nd_raw = np.std(ndr,axis=0)
med_nd_raw = np.median(ndr,axis=0)

def array2poppyhdul(array,pixscal):
    
    # OpticStudio units are in waves, poppy likes .fits files
    hdu_p =fits.PrimaryHDU(array)
    hdul_p = fits.HDUList([hdu_p])
    hdul_p[0].header['PIXELSCL'] = pixscal
    hdul_p[0].header['BUNIT'] = 'arcsec'
    
    return hdul_p

radius,std_wd_pr = poppy.radial_profile(array2poppyhdul(std_wd,0.0210804))
radius,med_wd_pr = poppy.radial_profile(array2poppyhdul(med_wd,0.0210804))
radius,std_nd_pr = poppy.radial_profile(array2poppyhdul(std_nd,0.0210804))
radius,med_nd_pr = poppy.radial_profile(array2poppyhdul(med_nd,0.0210804))
radius,std_nd_raw_pr = poppy.radial_profile(array2poppyhdul(std_nd_raw,0.0210804))
radius,med_nd_raw_pr = poppy.radial_profile(array2poppyhdul(med_nd_raw,0.0210804))
radius,std_wd_raw_pr = poppy.radial_profile(array2poppyhdul(std_wd_raw,0.0210804))
radius,med_wd_raw_pr = poppy.radial_profile(array2poppyhdul(med_wd_raw,0.0210804))
```
