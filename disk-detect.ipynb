# Convert processed disks to SNR
This is accomplished by finding the mean value in an annular region of the mean frame of each case (+11 & -11 *without* disks) to serve as the speckle noise

$$SNR = \frac{Frame}{mean(Annular frame w/o disk)}$$

```python
# convert disks to SNR

# Compute mean frame
mean_p11_nd = np.mean(p11box_nd,axis=0)
mean_m11_nd = np.mean(m11box_nd,axis=0)

# Grab mean in annulus
x = np.linspace(-1,1,67)
y = x
x,y = np.meshgrid(x,y)
r = np.sqrt(x**2 + y**2)

maskarray = (r<=0.75) & (r>= 0.2)

mean_an_p11_nd = np.mean(mean_p11_nd[maskarray])
mean_an_m11_nd = np.mean(mean_m11_nd[maskarray])

# convert boxes to SNR

p11box_nd *= 1/mean_an_p11_nd
p11box_disk *= 1/mean_an_p11_nd
m11box_nd *= 1/mean_an_m11_nd
m11box_disk *= 1/mean_an_m11_nd

# Now the raw data
for rind in range(len(rawbox[:])):
    rawbox[rind] *= 1/np.mean(rawbox[rind][maskarray])
#     plt.figure()
#     plt.imshow(rawbox[rind])
#     plt.colorbar()
#     plt.show()
```

# NMF Subtraction

```python
def NMFSub(data,components,mask=None,cube=False):
    # Use around 1/3 of available components
    numcomps = 100
    
    # Preallocate 
    resultbox = np.zeros(components.shape)
    results = np.zeros(data.shape)
    for i in range(data.shape[0]):
        trg = data[i]
        
        model = nmf_imaging.NMFmodelling(trg=trg,
                                        components=components,
                                        n_components=numcomps,
                                        maxiters=1e2,
                                        mask_components=mask,
                                        cube=cube)
        
        if cube==True:
            for l in range(numcomps):
                best_frac = nmf_imaging.NMFbff(trg,model[l])
                resultbox[l,:,:] += nmf_imaging.NMFsubtraction(trg,model[l],frac=best_frac)[0]
        else:
            best_frac = nmf_imaging.NMFbff(trg,model)
            results[i] = nmf_imaging.NMFsubtraction(trg,model,frac=best_frac)
    
    if cube==True:
        return resultbox
    
    else:
        return results
```

```python
def masknan(array,replace):
    """
    Set values NMF doesn't like to some value
    """
    array[np.isnan(array)] = replace
    array[array==np.inf] = replace
    array[array<0] = replace
    
    return array

m11box_disk = masknan(m11box_disk,1e-12)
p11box_disk = masknan(p11box_disk,1e-12)
m11box_nd = masknan(m11box_nd,1e-12)
p11box_nd = masknan(p11box_nd,1e-12)
components = masknan(components,1e-12)

# m11box_disk[m11box_disk <= 0] = 1e-12
# m11box_disk[m11box_disk == np.nan] = 1e-12
# m11box_disk[m11box_disk == np.inf] = 1e-12

# p11box_disk[p11box_disk <= 0] = 1e-12
# p11box_disk[p11box_disk == np.nan] = 1e-12
# p11box_disk[p11box_disk == np.inf] = 1e-12

# m11box_nd[m11box_nd <= 0] = 1e-12
# m11box_nd[m11box_nd == np.nan] = 1e-12
# m11box_nd[m11box_nd == np.inf] = 1e-12

# p11box_nd[p11box_nd <= 0] = 1e-12
# p11box_nd[p11box_nd == np.nan] = 1e-12
# p11box_nd[p11box_nd == np.inf] = 1e-12

# components[components <= 0] = 1e-12
# components[components == np.nan] = 1e-12
# components[components == np.inf] = 1e-12

m11_nmf_nc = NMFSub(m11box_disk,components,mask=mask,cube=False)
p11_nmf_nc = NMFSub(p11box_disk,components,mask=mask,cube=False)

m11_nmf_nd_nc = NMFSub(m11box_nd,components,mask=mask,cube=False)
p11_nmf_nd_nc = NMFSub(p11box_nd,components,mask=mask,cube=False)
```
